```python
import requests
import re
def parse_urls(urls):
    result_dict = {}
    # Обходим каждый URL в списке
    for url in urls:
        # Отправляем GET-запрос на сайт
        response = requests.get(url)

        # Используем регулярное выражение для поиска ссылки href в теге <a> с классом "mantine-UnstyledButton-root"
        href_pattern = r'<a class="mantine-UnstyledButton-root mantine-Button-root mantine-1movwzg" type="button" data-button="true" href="/api/download/models/(\d+)" download="">'
        href_match = re.search(href_pattern, response.text)

        # Если ссылка была найдена, извлекаем id и сохраняем его в словарь
        if href_match:
            href_id = href_match.group(1)
            # Извлекаем символьную последовательность из URL после последнего символа слэш
            key = url.rsplit('/', 1)[-1]
            result_dict[key] = href_id
        else:
            result_dict[key] = 'Ссылка href не найдена.'
    return result_dict

# Пример использования
urls_checkpts = ["https://civitai.com/models/24779/dark-sushi-mix-mix",


"https://civitai.com/models/43331/majicmix-realistic",

"https://civitai.com/models/57319/a-zovya-photoreal",

"https://civitai.com/models/7371/rev-animated",

"https://civitai.com/models/41865/majicmix-fantasy",

"https://civitai.com/models/30240/toonyou",

"https://civitai.com/models/47464/tang-yuan-mix",

"https://civitai.com/models/1265/analog-diffusion"]
urls_lora=["https://civitai.com/models/25995/blindbox",

"https://civitai.com/models/51686/glowingrunesai",

"https://civitai.com/models/48859/crystallineai",

"https://civitai.com/models/54524/on-fire",

"https://civitai.com/models/22359/betterrameneating",

"https://civitai.com/models/28857/flashlight-photography",

"https://civitai.com/models/48007/yoru-or-chainsaw-man",

"https://civitai.com/models/52652/instant-photo-polaroid-lora",

"https://civitai.com/models/33990/esthetic-makima-chainsaw-man"
]
result_dict_checkpoints = parse_urls(urls_checkpts)
print(result_dict_checkpoints)
result_dict_loras = parse_urls(urls_lora)
print(result_dict_loras)
```
